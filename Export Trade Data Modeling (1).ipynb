### Fetching all the Tables from Export Trade Database.

```python
query = "SHOW TABLES"
mycursor.execute(query)
table_list = mycursor.fetchall()  # Always fetch results
final_table_list = [i[0] for i in table_list]
final_table_list
```

----

### SQL Query to Perform Data Merge.

```python
final_query = ""

for index,table_name in enumerate(final_table_list):
    final_query += f"SELECT * FROM `export_trade_database`.`{table_name}` \n"

    if index!= len(final_table_list)-1:
        final_query += 'UNION ALL \n'
final_query += ';'
final_merging_query = final_query
print(final_merging_query)
```

### Performing Data Merge.

```python
start_time = time.time()
query = f"CREATE TABLE Country_wise_export_data AS {final_merging_query}"
mycursor.execute(query)
end_time = time.time()

print("Data Merged successfully.")
print(f"Time taken: {end_time - start_time:.2f} seconds")
```
