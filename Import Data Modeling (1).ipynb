----

### Fecting all the tables name in Import Trade Database.

```python
query = "SHOW TABLES"
mycursor.execute(query)
table_list = mycursor.fetchall()  # Always fetch results
final_table_list = [i[0] for i in table_list]
final_table_list
```

----

### SQL Query to Perform Data Merge.

```python
final_query = ""

for index,table_name in enumerate(final_table_list):
    final_query += f"SELECT * FROM `import_trade_database`.`{table_name}` \n"

    if index!= len(final_table_list)-1:
        final_query += 'UNION ALL \n'
final_query += ';'
final_merging_query = final_query
print(final_merging_query)
```

----

### Creating Database to store Merged Data File.

```python
query = 'CREATE DATABASE 003_Export_Potential_Index_Project'
mycursor.execute(query)
print("Database created successfully.")
```

```python
query = f"CREATE TABLE Country_wise_import_data AS {final_merging_query}"
mycursor.execute(query)
print("Data Merged successfully.")
```
